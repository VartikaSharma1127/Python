Operator's
An operator is a symbol that tells the interpreter to perform specific mathematical or logical manipulations. An operator require's values and operator to perform the operation.

Python has the following type of operators

1) Arthmetic Operators

2) Assignment Operators

3) Comparison Operators

4) Logical Operators

5) Identity Operators

6) Membership Operators

1) Airthmetic Operators

These are the operators that are used when performing mathematical operations.

```
#Addition
print(2+3)

#Subtraction
print(10 - 6)

#multiplication
print(2*8)

#modulus
print(9%2)

#division
print(5/2)

#Exponentiation
print(2**3)

#Floor Division
print(11//2)
```

2) Assignment Operators

These operators are used when assigning values.


```
#Addition Assignment
a = 2
a += 5 #Same as a = a + 5
print(a)

#Subtraction Assignment
a = 20
a -= 5 #Same as a = a - 5
print(a)

#Multiplication Assignment
a = 2
a *= 5 #Same as a = a * 5
print(a)


#Division Assignment
a = 20
a /= 5 #Same as a = a / 5
print(a)


#Modulas Assignment
a = 23
a %= 5 #Same as a = a % 5
print(a)

#Floor Division Assignment
a = 24
a //= 5 #Same as a = a // 5
print(a)

#Exponentiation Assignment
a = 2
a **= 3 #Same as a = a ** 3
print(a)
```

4) Logical Operators

These operators are used when we want to combine condition's with operators.

and

Returns true if both expressions return true.

```
2 > 3 and 2 == 2
```

```
2 < 3 and 2 == 2
```

or

Returns true if either expressions return true

```
2 > 3 or 5 < 2
```

```
2 > 3 or 2 == 2
```

not

Returns opposite of expression.

```
not(2 > 3 or 2 == 2)
```

```
not(2 < 3 and 5 < 2)
```

Write a Python program that allows the user to enter a four-digit binary number and displays its value in base 10. Each binary digit should be entered one per line, starting with the leftmost digit, as shown below.

DecimalNumber=ForthBit∗23+ThirdBit∗22+SecondBit∗21+FirstBit∗20

```
# program to convert a four bit binary numbr into decimal number 

firstbit = int(input("Enter the left most bit of number : "))
secondbit = int(input("Enter the second left most bit of number : "))
thirdbit = int(input("Enter the third left most bit of number : "))
forthbit = int(input("Enter the right most bit of number : "))

decimal_number = forthbit*2**3 + thirdbit*2**2 + secondbit*2**1 + firstbit*2**0

print("The equivalent decimal number is :", decimal_number)
```

```
binary = input("Enter binary number string:" )

power = len(binary) - 1
decimal = 0
for i in binary:
    decimal = decimal + int(i)*2**power
    power = power - 1

print("The equivalent decimal number is %d of binary string %s" %(decimal, binary))
```
